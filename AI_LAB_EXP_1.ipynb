```python
#IMPORT LIBRARIES

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import tensorflow as tf
import cv2
from sklearn.preprocessing import LabelEncoder
print("="*60)
print("INTEODUCTION TO PYTHON LIBRARIES FOR AI")
```

```python
# NUMPY

print("\n-----Numpy-----")

numbers=np.array([10 , 20 , 30 , 40 , 50])

print("Array:",numbers)
print("sum:",np.sum(numbers))
print("Mean:",np.mean(numbers))
print("Maximum:",np.max(numbers))
print("Minimum:",np.min(numbers))
```

```python
# PANDAS

print("\n------Pandas-----")

student={
    "name":["Utsav","jit","soumojit"],
    "Marks":[90,80,75]
}

df=pd.DataFrame(student)
print("\n Student Data")
print(df)
```

```python
# MATPLOTLIB

print("----Matplorlib-----")
student={
    "name":["Utsav","jit","soumojit"],
    "Marks":[90,80,75]
}

df=pd.DataFrame(student)
print("Displaying bar chart---")
plt.bar(df["name"],df["Marks"])
plt.title("student data")
plt.xlabel("name")
plt.ylabel("Marks")
plt.show()
```

```python
# TENSORFLOW

print("\n-----Tensoflow------")

tensor1=tf.constant([1,2,3])
tensor2 = tf.constant([4,5,6])

print("Tensor1:",tensor1.numpy())
print("Tensor2:",tensor2.numpy())

result=tf.add(tensor1,tensor2)
print(" Tensor Addition:",result.numpy())

```

```python
# OPEN CV

print("---- Open CV-----")

# create a blank black image
image =np.zeros((250,250,3),dtype=np.uint8)

#draw a blue regtangle

cv2.rectangle(image,(50,50),(200,200),(255,0,0),3)

#draw a green circle
cv2.circle(image,(125,125),40,(0,255,0),3)

#display image

plt.imshow(cv2.cvtColor(image,cv2.COLOR_BGR2RGB))
plt.title("open CV shapes")
plt.axis("off")
plt.show()
```
